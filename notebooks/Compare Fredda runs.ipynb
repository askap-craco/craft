```python
import numpy as np
from pylab import *
%matplotlib inline
from plot_fredda import load4d
```

```python
dir='/data/TETHYS_1/ban115/testdata/vela/SB12093/20200228033019/ICS/C000/testing'

def load2(fname, start=1):
    f1 = load4d(os.path.join(dir, 'run{}'.format(start), fname))
    f2 = load4d(os.path.join(dir, 'run{}'.format(start+1),fname))

    return (f1, f2)
```

```python
fdmt1, fdmt2 = load2('fdmt_e2.dat',15)
```

```python
np.all(fdmt1 == fdmt2)
```

```python
np.sum(fdmt1 != fdmt2)
```

```python
fig, axs = subplots(1,3)
for ibeam, ax in enumerate(axs.flatten()):
    d = fdmt1[ibeam,0,:,:] - fdmt2[ibeam,0,:,:]
    ax.imshow(d, aspect='auto', origin='lower')
```
