```python
%matplotlib inline
from pylab import *
import numpy as np
```

```python
frbs = ['180924','181112', '121102','171020']
redshift = np.array([0.321, 0.4749,0.19273,0.00867])
dm_frb = np.array([362.4, 589, 568,114])
dm_mw = np.array([40.5, 40.2, 188, 38])
dm_host = np.array([0,0,0,0])
dm_halo = 60


```

```python
dm_igm = dm_frb - dm_mw - dm_host - dm_halo
```

```python
scatter(redshift, dm_igm)
xlabel('z')
for ifrb, f in enumerate(frbs):
    text(redshift[ifrb], dm_igm[ifrb], 'FRB'+f)
ylabel('$\rm{DM_{IGM}}$ (pc/cc)')
zrange = np.arange(0, 0.7, 0.1)
predicteddm = zrange*1000
plot(zrange, predicteddm)
text(0.6, 1200*0.6, 'z*1000')
xlim(0,0.7)
ylim(0, 700)
```
